## Wizualiacja t-SNE

````
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
````

````
tokens = []
labels = []

for word in list(gwv.vocab.keys())[:1000]:
  tokens.append(gwv[word])
  labels.append(word)


tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500, random_state=23)
new_values = tsne_model.fit_transform(tokens)

x = []
y = []
for value in new_values:
    x.append(value[0])
    y.append(value[1])

plt.figure(figsize=(16, 16))
for i in range(len(x)):
    plt.scatter(x[i],y[i])
    plt.annotate(labels[i],
                  xy=(x[i], y[i]),
                  xytext=(5, 2),
                  textcoords='offset points',
                  ha='right',
                  va='bottom')
plt.show()
````

Dla pytania użytkownia pytania `q_user` wektoryzujemy je i liczmy podobieństwo z pytaniami w bazie

```python
sim("q_user_vectorized","q1_vecotorized"), sim("q_user_vectorized","q2_vecotorized"), ... ("q_user_vectorized""qn_vecotorized")
```

Zwracamy odpowiedź do pytania z bazy najbardziej podobnego do pytania użytkownika
$$
argmax(sim(q_{user}vectorized,q_{i}vecotorized))
$$

Do liczenia podobnieństma można użyć np https://docs.scipy.org/doc/scipy/reference/generated/scipy.spatial.distance.cosine.html ale wtedy robimy argmin bo
$$
cosine\ distance = 1 - cosine\ similarity
$$

````
def vectorize_question(question):
    words = question.split()
    word_vectors = [wv[word] for word in words if word in wv]
    return np.mean(word_vectors, axis=0)
````

````
qa_vectors = [{"vector": vectorize_question(qa["question"]), "answer": qa["answer"]} for qa in qa_pairs]
````

````
def get_user_query_and_find_answer():
    user_query = input("Please enter your question: ")
    user_vector = vectorize_question(user_query)

    def find_best_match(user_vector, qa_vectors):
        similarities = [(1 - cosine(user_vector, qa["vector"]), qa["answer"]) for qa in qa_vectors]
        best_match = max(similarities, key=lambda x: x[0])
        return best_match[1]

    answer = find_best_match(user_vector, qa_vectors)
    print("Answer:", answer)
````
